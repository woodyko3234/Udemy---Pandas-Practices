```python
import pandas as pd
pd.core.common.is_list_like = pd.api.types.is_list_like
from pandas_datareader import data

import matplotlib.pyplot as plt
%matplotlib inline
```

## The `.plot() ` Method

```python
bb = data.DataReader(name = "GOOG", data_source = "morningstar", 
                         start = "2010-01-01", end = "2017-12-31")
bb.head(3)
```

```python
bb.plot(y = 'High')
```

```python
bb[["High", "Low"]].plot()
```

## Bar Charts

```python
def rank_performance(stock_price):
    if stock_price <= 700:
        return "Poor"
    elif stock_price > 700 and stock_price <= 1000:
        return "Satisfactory"
    else:
        return "Stellar"
```

```python
plt.style.use("ggplot")
bb["Close"].apply(rank_performance).value_counts().plot(kind = "bar")
```

## Pie Charts

```python
apple = data.DataReader(name = "AAPL", data_source = "morningstar", start = "2012-01-01", end = "2016-12-31")
apple.head(3)
```

```python
apple["Close"].mean()
```

```python
def rank_performance(stock_price):
    if stock_price >= 92.63809405940599:
        return "Above Average"
    else:
        return "Below Average"
```

```python
plt.style.use("ggplot")
apple["Close"].apply(rank_performance).value_counts().plot(kind = "pie", legend = True)
```

## Histograms

```python
google = data.DataReader(name = "GOOG", data_source = "morningstar", start = "2004-01-01", end = "2016-12-31")
google.head(3)
```

```python
def custom_round(stock_price):
    return int(stock_price / 100.0) * 100
```

```python
google["High"].apply(custom_round).value_counts().sort_index()
```

```python
google["High"].apply(custom_round).nunique()
```

```python
google["High"].apply(custom_round).plot(kind = "hist", bins = 9)
```
