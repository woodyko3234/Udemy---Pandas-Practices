## Outer Joins

```python
merged = week1.merge(week2, how = "outer", on = "Customer ID", suffixes = [" - W1", " - W2"],
            indicator = True)
```

```python
merged.head()
```

```python
merged["_merge"].value_counts()
```

```python
mask = merged["_merge"].isin(["left_only", "right_only"])
merged[mask]
```

## Merging by Indexes with the `left_index` and `right_index` Parameters

Note: If we are going to merge different datasets on different column names or index, left_on(for column)/left_index and right_on/right_index are required.

```python
week1 = pd.read_csv("pandas/Restaurant - Week 1 Sales.csv")
week2 = pd.read_csv("pandas/Restaurant - Week 2 Sales.csv")
customers = pd.read_csv("pandas/Restaurant - Customers.csv", index_col = "ID")
foods = pd.read_csv("pandas/Restaurant - Foods.csv", index_col = "Food ID")
```

```python
foods.head(3)
```

```python
sales = week1.merge(customers, how = "left", left_on = "Customer ID", right_index = True)
sales = sales.merge(foods, how = "left", left_on = "Food ID", right_index = True)
sales.head(3)
```

```python
week1.head(3)
```

```python
week2.head(3)
```

```python
week1.merge(week2, how = "left", left_index = True, right_index = True, suffixes = [" - W1", " - W2"])
```

## The `.join()` Method

```python
satisfaction.head(3)
```

```python
week1.merge(satisfaction, how = "left", left_index = True, right_index = True).head(3)
```
